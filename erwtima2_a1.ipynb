```python
#read train set data from csv
import pandas as pd
from ast import literal_eval
trainSet = pd.read_csv(
'train_set.csv', # replace with the correct path
converters={"Trajectory": literal_eval},
index_col='tripId'
)
```

```python
#read test set data from csv, sep=\t
testSet = pd.read_csv(
'test_set_a1.csv',converters={"Trajectory": literal_eval},sep='\t') # replace with the correct path
```

```python
train_index_list=[]
train_index_list=trainSet.index
lenght_index_train=len(train_index_list)
```

```python
test_index_list=testSet.index.values
length_index_test=len(test_index_list)
```

```python
#to write distances with the ids
data={'id':[],
      'distances':[]}
df=pd.DataFrame(data,columns=['id','distances'])
row_counter=1
```

```python
from fastdtw import fastdtw
from haversine import haversine
import gmplot
import time
```

```python
#Test Trip 1
start = time.time()
#for test_0
pos_test=test_index_list[0]
traj_test=testSet['Trajectory'][0]

#list with map names
map_names_0=["a1_0_neighbor1.html","a1_0_neighbor2.html","a1_0_neighbor3.html","a1_0_neighbor4.html","a1_0_neighbor5.html"]

#create map for test_0
lons= []
lats= []
#all lons from train row
for i in range(1,len(traj_test)):
    lons.append(traj_test[i][1])
#all lats from train row
for i in range(1,len(traj_test)):
    lats.append(traj_test[i][2])


length=(len(lats))
gmap = gmplot.GoogleMapPlotter(lats[length/2], lons[length/2], 14)
gmap.plot(lats, lons, 'cornflowerblue', edge_width=10)
gmap.draw("a1_0_test0.html")
lats*=0
lons*=0
train_x =[]
test_y  =[]
for i in range(1,len(traj_test)):
    test_y.append(traj_test[i][1:])

#get distances for each row
for j in range(0,lenght_index_train):
    trip_id=train_index_list[j]
    traj_train=trainSet['Trajectory'][trip_id]
    for i in range(1,len(traj_train)):
        train_x.append(traj_train[i][1:])
        
    
    ###########
    distance, path = fastdtw(test_y,train_x, dist=haversine)
    train_x*=0
    df.ix[row_counter,'id']=trip_id
    df.ix[row_counter,'distances']=distance
    row_counter+=1
    

#new data = sorted distances
new_data=df.sort_values(by='distances')

#write sorted distances to csv
new_data.to_csv('test.csv',sep='\t',index=False)
test_data=pd.read_csv('test.csv',sep='\t')
for i in range(0,5):
    ids_0.append(test_data.iloc[i][0])
    distances_0.append(test_data.iloc[i][1])

#create map
for i in range (0,len(ids_0)):
    #get jp_id for neighbor
    jp_id_1=trainSet['journeyPatternId'][ids_0[i]]
    print jp_id_1
    print distances_0[i]
    map_name=map_names_0[i]
    #create map for neighbor 1
    traj=trainSet['Trajectory'][ids_0[i]]
    lons= []
    lats= []
    #all lons from train row
    for k in range(1,len(traj)):
        lons.append(traj[k][1])
    #all lats from train row
    for l in range(1,len(traj)):
        lats.append(traj[l][2])
    length=(len(lats))
    gmap = gmplot.GoogleMapPlotter(lats[length/2], lons[length/2], 14)
    gmap.plot(lats, lons, 'cornflowerblue', edge_width=10)
    gmap.draw(map_name)
    lats*=0
    lons*=0
    
#clear for next one
pos_test=0
#traj_test*=0
#traj_train*=0
train_x*=0
test_y*=0
#ids*=0
row_counter=1
distance*=0
path*=0

end = time.time()
time_for_0= end - start
print time_for_0
```
